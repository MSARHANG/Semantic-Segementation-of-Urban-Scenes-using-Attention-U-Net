```python
import os
import dataset
import model
import train
import utils
import configs
```

```python
image_list, mask_list = dataset.load_image_paths(configs.IMAGE_DIR, configs.MASK_DIR)
dataset = dataset.prepare_dataset(image_list, mask_list, configs.BATCH_SIZE, configs.BUFFER_SIZE)
```

```python
model = model.attention_unet(input_shape=(96, 128, 3), num_classes=configs.NUM_CLASSES)
```

```python
history = train.compile_and_train(model, dataset, 6, configs.BUFFER_SIZE, configs.BATCH_SIZE)
```

```python
import matplotlib.pyplot as plt

def plot_predictions(model, dataset, num=1):
    for image, mask in dataset.take(num):
        pred_mask = model.predict(image)
        pred_mask_processed = train.create_mask(pred_mask)
        
        plt.figure(figsize=(12, 4))

        # Plot original image
        plt.subplot(1, 3, 1)
        plt.title("Input Image")
        plt.imshow(image[0])
        plt.axis('off')

        # Plot actual mask
        plt.subplot(1, 3, 2)
        plt.title("Actual Mask")
        plt.imshow(mask[0])
        plt.axis('off')

        # Plot predicted mask
        plt.subplot(1, 3, 3)
        plt.title("Predicted Mask")
        plt.imshow(pred_mask_processed)
        plt.axis('off')

        plt.show()
```

```python
plot_predictions(model, dataset, num=5)
```
